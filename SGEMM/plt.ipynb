```python
import matplotlib.pyplot as plt

# 示例数据（替换为你的实验数据）
N = [128, 256, 384, 512, 640, 768, 896, 1024, 1536, 2048, 4096]
naive      = [301.75, 585.67, 664.13, 693.62, 713.13, 725.52, 723.09, 728.13, 721.14, 712.78, 337.44]
shared     = [254.22, 653.98, 825.49, 899.28, 918.82, 944.49, 951.54, 956.47, 965.93, 992.76, 692.87]
wmma       = [352.54, 1581.32, 3349.71, 4112.12, 4361.41, 4849.91, 5014.64, 5065.91, 5319.02, 5394.66, 5162.68]
wmma_smem  = [230.56, 1028.78, 2339.84, 2969.14, 3139.66, 4015.97, 3785.04, 4397.83, 4484.73, 4676.45, 4858.52]
pipeline   = [498.36, 1834.90, 4861.70, 6693.81, 7291.19, 8709.40, 9192.74, 9320.36, 10014.92, 10150.10, 7793.40]

# 等间距位置
x = list(range(len(N)))

# 绘制图形
plt.figure(figsize=(6,4))

plt.plot(x, naive,     marker='o', label='Naive CUDA')
plt.plot(x, shared,    marker='s', label='Shared Memory (CUDA Core)')
plt.plot(x, wmma,      marker='^', label='WMMA (Tensor Core)')
plt.plot(x, wmma_smem, marker='x', label='WMMA + Simple Shared Memory')
plt.plot(x, pipeline,  marker='d', label='WMMA + Pipeline (Double Buffer)')

plt.xlabel('Matrix Size (N)')
plt.ylabel('GFLOPS')
plt.xticks(x, N, rotation=0)  # 用真实 N 做刻度标签

# 调整 legend 字体大小
plt.legend(loc='upper left', fontsize=9, frameon=True)
plt.grid(True)
plt.tight_layout()

plt.savefig('gpu_sgemm_performance.pdf')
plt.show()
```
